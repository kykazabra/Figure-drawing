```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib as mpl
# Импортируем все необходимые модули
```

### Ввод для всех типов фигур одинаковый. Угол вводится в градусах, смещение может быть любым численным, симметрия - (да, 1) или (нет, 0).

```python
def gen_rectangle(amount, a, b, degree, move, symmetry): # Метод для создания прямоугольников
    xm, ym = move # Координаты для смещения ряда 
    ax = fig.add_subplot(111) # Генерация "фона"
    for i in range(amount):
        ax.add_patch(patches.Rectangle(((a + 1)*i + xm, 0 + ym), a, b, angle = degree * symmetry, fc = '#008B8B'))
        # Цикличное создание новых прямоугольников
        
    xlim = a * amount
    ylim = b * amount // 2 
    plt.plot(xlim, ylim)
    # Настраиваем координатные оси для соразмерного отображения
    
fig = plt.figure(figsize = (16, 5)) # Настраиваем масштаб поля отображения
degree = int(input('Введите угол поворота ')) # Пользователь вводит угол поворота фигуры
move = (int(input('Введите сдвиг по оси X ')), int(input('Введите сдвиг по оси Y '))) 
# Пользователь вводит смещение по OX и OY
answer = input('Использовать симметрию? ') # Пользователь решает, нужна ли симметрия
if answer.lower() in ('да', '1'):
    symmetry = 1
if answer.lower() in ('нет', '0'):
    symmetry = -1 
gen_rectangle(7, 2, 1, degree, move, symmetry) # Вызываем создание прямоугольников со всеми заданыыми параметрами
plt.show() # Отрисовываем получившееся поле
```

###  По умолчанию задается 7 прямоугольников 2 х 1, выглядят как квадраты из за динамечиеских осей. Симметрию я в данном случае реализовал как смену угла поворота фигуры на обратный.

```python
def gen_triangle(amount, a, degree, move, symmetry): # Метод для создания треугольников
    xm, ym = move # Координаты для смещения ряда
    ax = fig.add_subplot(111) # Генерация "фона"
    for i in range(amount):
        ax.add_patch(patches.RegularPolygon((a + 2*i*a + xm, a + ym), numVertices = 3, radius = a, 
                                            orientation = degree * symmetry, fc = '#808000'))
        # Цикличное создание новых треугольников
        
    xlim = a * amount * 2
    ylim = 3 * a
    plt.plot(xlim, ylim)
    # Настраиваем координатные оси для соразмерного отображения
    
fig = plt.figure(figsize = (16, 5)) # Настраиваем масштаб поля отображения
degree = int(input('Введите угол поворота ')) # Пользователь вводит угол поворота фигуры
move = (int(input('Введите сдвиг по оси X ')), int(input('Введите сдвиг по оси Y ')))
# Пользователь вводит смещение по OX и OY
answer = input('Использовать симметрию? ') # Пользователь решает, нужна ли симметрия
if answer.lower() in ('да', '1'):
    symmetry = 1
if answer.lower() in ('нет', '0'):
    symmetry = -1 
gen_triangle(7, 1, degree, move, symmetry) # Вызываем создание треугольников со всеми заданыыми параметрами 
plt.show() # Отрисовываем получившееся поле
```

###  По умолчанию задается 7 треугольников со стороной 1.

```python
def gen_hexagon(amount, a, degree, move, symmetry): # Метод для создания шестиугольников
    xm, ym = move # Координаты для смещения ряда
    ax = fig.add_subplot(111) # Генерация "фона"
    for i in range(amount):
        ax.add_patch(patches.RegularPolygon((a + 2*i*a + xm, a + ym), numVertices = 6, radius = a, 
                                            orientation = degree  * symmetry, fc = '#F08080'))
        # Цикличное создание новых шестиугольников
        
    xlim = 2 * amount * a 
    ylim = 3 * a
    plt.plot(xlim, ylim)
    # Настраиваем координатные оси для соразмерного отображения
    
fig = plt.figure(figsize = (16, 5)) # Настраиваем масштаб поля отображения
degree = int(input('Введите угол поворота ')) # Пользователь вводит угол поворота фигуры
move = (int(input('Введите сдвиг по оси X ')), int(input('Введите сдвиг по оси Y ')))
# Пользователь вводит смещение по OX и OY
answer = input('Использовать симметрию? ') # Пользователь решает, нужна ли симметрия
if answer.lower() in ('да', '1'):
    symmetry = 1
if answer.lower() in ('нет', '0'):
    symmetry = -1 
gen_hexagon(7, 2, degree, move, symmetry) # Вызываем создание шестиугольников со всеми заданыыми параметрами
plt.show() # Отрисовываем получившееся поле
```

```python
rng = np.linspace(0, 5, 8) # Размещаем 8 фигур на диапазоне от 0 до 5
plt.figure(figsize = (5, 5)) # Настраиваем масштаб поля отображения

plt.plot(rng, rng, linestyle = '', marker = 'H', markersize = 20, markeredgecolor = '000000', markerfacecolor = '#2EF9F2')
plt.plot(rng, rng + 1, linestyle = '', marker = 'H', markersize = 20, markeredgecolor = '000000', 
         markerfacecolor = '#2EF9B5')
plt.plot(rng, rng - 1, linestyle = '', marker = 'H', markersize = 20, markeredgecolor = '000000', 
         markerfacecolor = '#EBF92E')
# Создаем последовательности полигонов

plt.axis([-0.4, 5.4, -1.8, 6.6]) # Настраиваем оси 
plt.show() # Отрисовываем получившееся поле
```
